```python
#We will calculate first the average of goals in one game for each player. 
goal_game= data["Goals"]/data["Appearences"] #Average variable called "goal_game"
data["Average Goals per game"]= goal_game  #New column called "Average Goals per game with the values of goal_game
```

```python
print("Top 5 players with best average goals:")
best_players = data.set_index("Players")["Average Goals per game"].sort_values(ascending = False) #We say that Players will be always the index in the variable "best_players", and we sort the values from "Average Goals per game" from biggest to lowest.
bp = best_players.head()
best_players.head()
```

```python
#Bar Graph Best Average Goals
bp.plot(kind="bar", figsize=(3, 5), legend=False)
plt.title("Players with best average goals")
plt.xlabel("Players")
plt.ylabel("Average Goals")
plt.xticks(rotation= 75)
plt.savefig("players_average_goals.png")
plt.show
```

```python
#Which team has the best average in goals per game?
print("Top 5 teams with best average goals:")
group1 = data.groupby("Team")["Average Goals per game"].mean().sort_values(ascending = False)# We make a group with "Team" and "Average Goals per game", and it will return the average from biggest to lowest. 
g1 = group1.head()
group1.head()
```

```python
#Bar Graph Best Average Goals
g1.plot(kind="bar", figsize=(3, 5), legend=False)
plt.title("Teams with best average goals")
plt.xlabel("Teams")
plt.ylabel("Average Goals")
plt.xticks(rotation= 75)
plt.savefig("teams_average_goals.png")
plt.show
```

```python
#Which player has the best Shooting Precision?
data["Shooting Precision"]=data["Shots on target"]/data["Shots"] 
print("Top 5 players with the best Shooting precision")
best_players_1 = data.set_index("Players")["Shooting Precision"].sort_values(ascending=False)#We say that Players will be always the index in the variable "best_players_1", and we sort the values from "Shooting Precision" from biggest to lowest.
bp1 = best_players_1.head()
bp1
```

```python
#Bar Graph Best teams shooting precision
bp1.plot(kind="bar", figsize=(3, 5), legend=False, color = "green")
plt.title("Players with highest shooting precision")
plt.xlabel("Players")
plt.ylabel("Shooting precision")
plt.xticks(rotation= 75)
plt.savefig("players_shooting_precision.png")
plt.show
```

```python
#And which team has the best Shooting Precision?
print("Top 5 teams with the best Shooting Precision:")
group_2 = data.groupby("Team")["Shooting Precision"].mean().sort_values(ascending =False)# We make a group with "Team" and "Shooting Precision", and it will return the average from biggest to lowest.
g2 = group_2.head()
group_2.head()
```
